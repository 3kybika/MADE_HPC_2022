```python
import imageio as iio
import numpy as np
import skimage.color
import skimage.util
import matplotlib.pyplot as plt
```

```python
# read the image of a plant seedling as grayscale from the outset
image = iio.imread(uri="cosmos.jpg", pilmode="L")
```

```python
# convert the image to float dtype with a value range from 0 to 1
image = skimage.util.img_as_float(image)

# display the image
fig = plt.figure(figsize=(10, 10))
plt.imshow(image, cmap="gray")
```

```python
# read original image, in full color
image = iio.imread("cosmos.jpg")

# display the image
fig = plt.figure(figsize=(10, 10))
plt.imshow(image)
```

```python
# tuple to select colors of each channel line
colors = ("red", "green", "blue")

# create the histogram plot, with three lines, one for
# each color
plt.figure(figsize=(10, 10))
plt.xlim([0, 256])
for channel_id, color in enumerate(colors):
    histogram, bin_edges = np.histogram(
        image[:, :, channel_id], bins=256, range=(0, 256)
    )
    plt.plot(bin_edges[0:-1], histogram, color=color)

plt.title("Color Histogram (unnormalized)")
plt.xlabel("Color value")
plt.ylabel("Pixel count")
```
