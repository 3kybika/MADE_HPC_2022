Хотелось бы иметь топ компьютеров за более длительный промежуток времени, чем за последнее десятилетие:

```python
DOWNLOAD_REQUIRED = False

SIGN_IN_URL = "https://top500.org/accounts/login/"
DOWNLOAD_URL = "https://top500.org/lists/top500/{year}/{month}/download/TOP500_{year}{month}_all.xml"
FILE_NAME = "TOP500_{year}{month}.xml"
DATA_DIR_PATH = "./data/"
TOP_BY_YEARS_PATH = os.path.join(DATA_DIR_PATH, "top_by_years.csv")
CREDENTIALS_PATH = "credentials.json"
```

```python
# Уже подготовленный файлик есть в репозитории, так что нет необходимости в выполнении загрузки
if DOWNLOAD_REQUIRED:


    import json
    import requests
    from bs4 import BeautifulSoup
    
    with requests.Session() as session:

        # get credentials
        with open(CREDENTIALS_PATH) as credentials_file:
            credentials = json.load(credentials_file)
        
        # get csrf
        login_response = session.get(SIGN_IN_URL)
        soup = BeautifulSoup(login_response.text, "html.parser")
        csrf_token = soup.find_all("input")[0]["value"]
        
        # login
        data = {
            "login": credentials["login"],
            "password": credentials["password"],
            "csrfmiddlewaretoken": csrf_token
        }
        headers = {
            "Referer": SIGN_IN_URL
        }
        r = session.post(SIGN_IN_URL, data=data, headers=headers)
        
        # create data dir
        if not os.path.isdir(DATA_DIR_PATH):
            os.mkdir(DATA_DIR_PATH)

        # load data
        for year in range(1993, 2023):
            for month in ["06", "11"]: # site provides data only for june and november months
                file_name = os.path.join(DATA_DIR_PATH, FILE_NAME.format(year=year, month=month))
                download_url = DOWNLOAD_URL.format(year=year, month=month)
                
                # checking file exists:
                if os.path.isfile(file_name):
                    print("Already loaded: {}".format(file_name))
                    continue
                
                # load and write
                try:
                    file = session.get(download_url)
                    file.raise_for_status()

                    with open(file_name, 'wb') as f:
                        f.write(file.content)
                    print("Loaded: {}".format(file_name))
                except ConnectionRefusedError as err:
                    print("failed to load {} due to Connection error: {}".format(file_name, err))
                    pass
                except requests.exceptions.HTTPError as err:
                    print("failed to load {} due to HTTPError error: {}".format(file_name, err))
                    pass
```

Соберем данные в один датафрейм:

```python
top_by_years_df = None
if not os.path.isfile(TOP_BY_YEARS_PATH):
    # aggregate data
    def data_gen():
        for year in range(1993, 2023):
            for month in ["06", "11"]: # site provides data only for june and november months
                file_path = os.path.join(DATA_DIR_PATH, FILE_NAME.format(year=year, month=month))
                date = "{}-{}-01".format(year, month)

                # read data
                if not os.path.isfile(file_path):
                    continue
                df = pd.read_xml(file_path).set_index('rank')


                # extract
                yield {
                    "date": date,
                    "first_by_perf": df['r-max'].iloc[0],
                    "last_by_perf": df['r-max'].iloc[-1],
                    "sum_by_perf": df['r-max'].sum()
                }

    # write data
    top_df = pd.DataFrame.from_records(row for row in data_gen())
    top_df.to_csv(TOP_BY_YEARS_PATH)
else:
    top_df = pd.read_csv(TOP_BY_YEARS_PATH) 

top_df.head(100)
```

Соберем данные в один датафрейм:

```python
plt.figure(figsize=(20, 10))

plt.plot(top_df["date"], top_df["sum_by_perf"], "go", label='Sum of 500 [Flop/s]')
plt.plot(top_df["date"], top_df["first_by_perf"], "r^", label='#1 [Flop/s]')
plt.plot(top_df["date"], top_df["last_by_perf"], "bs", label='#500 [Flop/s]')

plt.title('Performance Development')
plt.ylabel('Flop/s')
plt.yscale('log')
plt.xticks(top_df['date'], rotation=30)

plt.grid(True)
plt.legend()

plt.show()
```

```python
def dt_to_int(dt):
    return int(datetime.strptime(dt, "%Y-%m-%d").timestamp())
dt_to_int_v = np.vectorize(dt_to_int)

def int_to_dt(dt):
    return datetime.strftime(datetime.fromtimestamp(dt), "%Y-%m-%d")
int_to_dt_v = np.vectorize(int_to_dt)
```

```python
pred_dates = np.array( ['2022-11-01'] + [
    "{}-{}-01".format(year, month)
    for year in range(2023, 2026) 
    for month in ["06", "11"]
]).astype(dtype=object)
pred_dates
```

Насколько известно вычислительная мощность подчиняется закону Мура - соответственно, мы можем использовать обычную линейную регрессию для предсказания:

```python
# prepare data
x = dt_to_int_v(top_df["date"].to_numpy())
x_pred = dt_to_int_v(pred_dates)
x_all = np.concatenate((x, x_pred))
all_dts = np.concatenate((top_df["date"].to_numpy(), pred_dates))

y_first = top_df["first_by_perf"]
y_last = top_df["last_by_perf"]
y_sum = top_df["sum_by_perf"]

def apply_polynom(koefs, args):
    res = koefs[-1]
    for i, k in enumerate(koefs[:-1]):
        res += k * np.power(args, len(koefs) - i -1)
    return res

def mse(Y, YH):
     return np.square(Y - YH).mean()

# training
koefs_sum = np.polyfit(x, np.log(y_sum), 2)
y_pred_sum = np.exp(apply_polynom(koefs_sum, x_pred))

koefs_first = np.polyfit(x, np.log(y_first), 2)
y_pred_first = np.exp(apply_polynom(koefs_first, x_pred))

koefs_last = np.polyfit(x, np.log(y_last), 2)
y_pred_last =  np.exp(apply_polynom(koefs_last, x_pred))
```

# Итог:

```python
plt.figure(figsize=(20, 10))

plt.plot(x,      y_sum,                                    "go", label='Sum of 500 [Flop/s]')
plt.plot(x_pred, y_pred_sum,                               "yo", label='Predict sum of 500 [Flop/s]')
plt.plot(x_all,  np.exp(apply_polynom(koefs_sum, x_all)),  "g:", label='Predict sum of 500 [Glop/s]')

plt.plot(x,      y_first,                                   "r^", label='#1 [Flop/s]')
plt.plot(x_pred, y_pred_first,                              "m^", label='Predict #1 [Flop/s]')
plt.plot(x_all,  np.exp(apply_polynom(koefs_first, x_all)), "r:", label='Predict sum of 500 [Flop/s]')

plt.plot(x,      y_last,                                    "bs", label='#500 [Flop/s]')
plt.plot(x_pred, y_pred_last,                               "cs", label='Predict #500 [Flop/s]')
plt.plot(x_all,  np.exp(apply_polynom(koefs_last, x_all)),  "b:", label='Predict sum of 500 [Flop/s]')

plt.title('Performance Development')
plt.ylabel('GFlop/s')
plt.yscale('log')
plt.xticks(
    ticks=dt_to_int_v(all_dts), 
    labels=all_dts, 
    rotation=30
)

plt.grid(True)
plt.legend()

plt.show()
```
