# Задача 2
Oцените какой процент всего производимого на Земле электричества потребляют все компьютеры из top500 (на основании открытых данных top500.org)

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn import linear_model
from datetime import datetime
```

Согласно: https://ru.wikipedia.org/wiki/Список_стран_по_производству_электроэнергии Всего в мире за 2021й год было выработано 28466,3 тераватт-час

```python
WORLD_PRODUCTUON = 28466.3 * 1e12
```

```python
src_df = pd.read_excel("green500_top_202206.xlsx")
src_df.head(500)
```

```python
src_df.info()
```

Будем исходить из предположения, что в течении года компьютеры работают не на максимальной, а на средней производительсноти.

```python
# fill empty values
src_df['Power (kW)'] = src_df['Power (kW)'].fillna(
    src_df['Power (kW)'].mean()
)

consumption = src_df['Power (kW)'].sum()
```

# Ответ:

```python
print("Энергия потребляемая суперкомпьютерами: {consumption:1.3f} kW".format(consumption = consumption))
print("Часть энергии, потребляемой суперкомпьютерами от общего количества выработанной энергии: {part: 1.6f} %".format(
    part=(consumption * 1e3 * 24 * 365 / WORLD_PRODUCTUON) * 100
))
```
